<h2>IMPORTING THE REQUIRED LIBRARIES</h2>

```python
import getpass
from sqlalchemy import create_engine
import pandas as pd
```

<h2>CREATING THE CONNECTION BETWEEN PYTHON AND ONE DATABASE FROM MYSQL</h2>

```python
def mysql_database_connection(database):

    password = getpass.getpass()
    connection = f'mysql+pymysql://root:{password}@localhost/{database}'
    engine = create_engine(connection)

    return engine

sakila_engine = mysql_database_connection('sakila')
```

<h2>EXERCISE 1</h2>

```python
query_1 = """SELECT COUNT(DISTINCT last_name) FROM actor"""

print(f'\nTOTAL NUMBER OF UNIQUE LAST NAMES: {pd.read_sql_query(query_1, sakila_engine).values[0][0]}')
```

<h2>EXERCISE 2</h2>

```python
query_2 = """SELECT *, CASE
             WHEN WEEKDAY(CONVERT(rental_date, DATE)) > 05 THEN 'WEEKEND'
             ELSE 'WORKDAY' END AS 'day_type'
             FROM rental"""

pd.read_sql_query(query_2, sakila_engine)
```

<h2>EXERCISE 3</h2>

<h2>EXERCISE 6</h2>

```python
query_6 = """SELECT rating, AVG(length) AS mean_duration FROM film
             GROUP BY rating
             HAVING mean_duration > 120"""

pd.read_sql_query(query_6, sakila_engine)
```

<h2>EXERCISE 7</h2>

```python
query_7 = """SELECT title, length,

             CASE
                WHEN length > 100 THEN 'long'
                WHEN length > 60 THEN 'medium'
                ELSE 'short'
             END AS length_rank

             FROM film

             WHERE length IS NOT NULL AND length <> 0
             ORDER BY length DESC"""

pd.read_sql_query(query_7, sakila_engine)
```
